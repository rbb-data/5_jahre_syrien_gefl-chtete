```python
import pandas as pd
import altair as alt
```

```python
data=pd.read_excel("../RawData/*auslaender_deutsche_gemeindeebene/table_2020-07-14_13-44-21.xlsx",sheet_name="Auslaender")
data["AGS"]="120"+data["Region"].str[:5]
data.iloc[0,-1]="11000000"
data
```

```python
Auslaender_df=pd.melt(data, id_vars=['Region',"AGS"], var_name='date', value_name='Ausländer',value_vars=["31.12.2018","31.12.2017","31.12.2016","31.12.2014","31.12.2015","31.12.2013"])
```

```python
data=pd.read_excel("../RawData/*auslaender_deutsche_gemeindeebene/table_2020-07-14_13-44-21.xlsx",sheet_name="Deutsche")
data["AGS"]="120"+data["Region"].str[:5]
data.iloc[0,-1]="11000000"
data
```

```python
Deutsche_df=pd.melt(data, id_vars=['Region',"AGS"], var_name='date', value_name='Deutsche',value_vars=["31.12.2018","31.12.2017","31.12.2016","31.12.2014","31.12.2015","31.12.2013"])
Deutsche_df
```

```python
auslaender_deutsche_df=pd.merge(Deutsche_df,Auslaender_df,on=["Region","AGS","date"])
auslaender_deutsche_df["Ausländer"]=pd.to_numeric(auslaender_deutsche_df["Ausländer"],errors="coerce")
auslaender_deutsche_df["Deutsche"]=pd.to_numeric(auslaender_deutsche_df["Deutsche"],errors="coerce")
auslaender_deutsche_df["Auslaender_Anteil"]=auslaender_deutsche_df["Ausländer"]/(auslaender_deutsche_df["Ausländer"]+auslaender_deutsche_df["Deutsche"])
auslaender_deutsche_df["date"]=pd.to_datetime(auslaender_deutsche_df["date"],format='%d.%m.%Y')
auslaender_deutsche_df
```

```python
for i in auslaender_deutsche_df["Region"].unique().tolist():
    df=auslaender_deutsche_df[auslaender_deutsche_df["Region"]==i]
    print(i)
    chart=alt.Chart(df).mark_line().encode(
    x='date',
    y="Auslaender_Anteil"
)
    chart.display()
```

```python
auslaender_deutsche_df.to_csv("../CleanedData/auslaender_deutsche_gemeinde.csv",index=False)
```
