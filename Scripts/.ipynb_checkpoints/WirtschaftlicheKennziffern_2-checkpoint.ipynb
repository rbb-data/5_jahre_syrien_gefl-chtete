```python
df=pd.read_excel("../RawData/wirtschaftliche_kennziffern_kreisebene/R2B1.xlsx",sheet_name=sheet_names[10])
new_df=df
category=df.iloc[0,0]
new_df.columns=df.iloc[3]
new_df=df.iloc[4:]
new_df=new_df.dropna(subset=["Regional-schlüssel"])
new_df
```

```python
data=pd.melt(new_df, id_vars=(new_df.columns.tolist()[:8]), value_vars=(new_df.columns.tolist()[8:]),
        var_name="years", value_name=category)
```

```python
based=data
based
```

```python
for index in sheet_names[11:]:
    print(index)
    df=pd.read_excel("../RawData/wirtschaftliche_kennziffern_kreisebene/R2B1.xlsx",sheet_name=index)
    new_df=df
    category=df.iloc[0,0]
    new_df.columns=df.iloc[3]
    new_df=df.iloc[4:]
    new_df=new_df.dropna(subset=["Regional-schlüssel"])
    data=pd.melt(new_df, id_vars=(new_df.columns.tolist()[:8]), value_vars=(new_df.columns.tolist()[8:]),
        var_name="years", value_name=category)
    based=pd.merge(based,data,on=based.columns.tolist()[:9])
based
```
