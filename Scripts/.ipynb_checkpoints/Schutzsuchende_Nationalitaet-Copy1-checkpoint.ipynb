```python
import pandas as pd
import glob
import altair as alt
```

```python
data=pd.read_csv("../RawData/nationalitaeten_aufenthaltstitel_kreis/status2018.csv",sep=";",dtype="str", encoding="latin3",skiprows=5)
data
```

```python
upperlist=data.iloc[0]
lowerlist=data.iloc[1]
columnlist=(upperlist+"_"+lowerlist).to_list()
columnlist[0]="Stichtag"
columnlist[1]="AGS"
columnlist[2]="Stadtname"
columnlist[3]="Herkunft"
data.columns=columnlist
data=data.dropna(subset=["AGS"])
data
```

```python
filenames=glob.glob("../RawData/schutzsuchende_nationalitaet_kreisebene/*")
filenames[:-1]
```

```python
basedata=data
for file in filenames[:-1]:
    data=pd.read_csv(file,sep=";",dtype="str", encoding="latin3",skiprows=5)
    upperlist=data.iloc[0]
    lowerlist=data.iloc[1]
    columnlist=(upperlist+"_"+lowerlist).to_list()
    columnlist[0]="Stichtag"
    columnlist[1]="AGS"
    columnlist[2]="Stadtname"
    columnlist[3]="Herkunft"
    data.columns=columnlist
    data=data.dropna(subset=["AGS"])
    basedata=pd.concat([basedata,data])
basedata
    
```

```python
for j in basedata.columns[4:].tolist():
    basedata[j]=pd.to_numeric(basedata[j],errors="coerce")
basedata
```

```python
basedata["Schutzsuchende_alle_Status"]=basedata[basedata.columns[4:].tolist()].sum(axis=1)
basedata
```

```python
basedata.to_csv("../CleanedData/schutzsuchende_nations_kreis.csv",index=False)
```

```python
onlybb_be=basedata[(basedata["AGS"].str.startswith("12"))|(basedata["AGS"].str.startswith("11"))]
onlybb_be
```

```python
sumover_rbb=onlybb_be.groupby(["Stichtag","Herkunft"]).sum().reset_index()
sumover_rbb
```

```python
for tag in sumover_rbb["Stichtag"].unique().tolist():
    year_df=sumover_rbb[sumover_rbb["Stichtag"]==tag]
    reduced_df=(year_df.sort_values(by=["Schutzsuchende_alle_Status"],ascending=False).head())
    print(reduced_df[["Herkunft","Schutzsuchende_alle_Status"]])
    chart=alt.Chart(reduced_df).mark_bar().encode(
    x='Herkunft:N',
    y="Schutzsuchende_alle_Status"
    )
    chart.display()
```

```python
chart=alt.Chart(sumover_rbb).mark_bar().encode(
x=alt.X('Herkunft:N', sort='-y'),
y="Schutzsuchende_alle_Status"
).facet(column="Stichtag")
chart.display()
```

```python
years2013_2018=sumover_rbb.groupby("Herkunft").sum().reset_index()
years2013_2018
```

```python
chart=alt.Chart(years2013_2018.sort_values(by="Schutzsuchende_alle_Status", ascending=False).head(10)).mark_bar().encode(
x=alt.X('Herkunft:N', sort='-y'),
y="Schutzsuchende_alle_Status"
)
chart.display()
```
