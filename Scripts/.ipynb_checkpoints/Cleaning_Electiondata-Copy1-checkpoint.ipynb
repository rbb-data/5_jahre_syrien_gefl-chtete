```python
import pandas as pd
```

```python
data=pd.read_excel("../RawData/DL_BB_2_LT2019 (1).xlsx", sheet_name="Brandenburg_Landtagswahl_A_2")
data=data.rename(columns={"Unnamed: 3":"Name"})
obereliste=data.columns.tolist()
untereliste=data.iloc[0].tolist()
untereliste=pd.Series(untereliste).fillna("").tolist()
data[data.columns[2]].unique()
```

```python
newlist=[]
for index in range(len(obereliste)):
    if str(obereliste[index]).startswith("Unnamed"):
        element=str(obereliste[index-1])+"_"+str(untereliste[index])
    else:
        element=str(obereliste[index])+"_"+str(untereliste[index])
    newlist.append(element)
data.columns=newlist
gemeinde_data_2019=data[(data[data.columns[2]]=='amtsfreie Gemeinde')| (data[data.columns[2]]=='amtsangehörige Gemeinde')]
gemeinde_data_2019["AGS"]=gemeinde_data_2019[gemeinde_data_2019.columns[1]].str[:8]
gemeinde_data_2019
```

```python
data=pd.read_excel("../RawData/BB_LT_2014_Aemter_Endg_Ergebnis (3).xlsx", sheet_name="Zweitstimmen_Gemeinden amtsfrei",skiprows=6, dtype=str)
data=data.rename(columns={"Unnamed: 0":"Gemeindenummer","Unnamed: 1":"Gemeindename","Unnamed: 6":"Wähler_Anzahl"})
obereliste=data.columns.tolist()
untereliste=data.iloc[0].tolist()
untereliste=pd.Series(untereliste).fillna("").tolist()
newlist=[]
for index in range(len(obereliste)):
    if str(obereliste[index]).startswith("Unnamed"):
        element=str(obereliste[index-1])+"_"+str(untereliste[index])
    else:
        element=str(obereliste[index])+"_"+str(untereliste[index])
    newlist.append(element)
data.columns=newlist
data
```

```python
gemeinde_data_2014_amtsfrei=data.dropna(subset=["Gemeindenummer_"])
gemeinde_data_2014_amtsfrei["AGS"]="120"+gemeinde_data_2014_amtsfrei["Gemeindenummer_"]
gemeinde_data_2014_amtsfrei
```

```python
gemeinde_data_2014_amtsfrei.columns
```

```python
data=pd.read_excel("../RawData/BB_LT_2014_Aemter_Endg_Ergebnis (3).xlsx", sheet_name="Zweitstimmen_Gemeinden amtsang.",skiprows=6, dtype="str")
data=data.rename(columns={"Unnamed: 0":"Gemeindenummer","Unnamed: 1":"Gemeindename","Unnamed: 6":"Wähler_Anzahl"})
obereliste=data.columns.tolist()
untereliste=data.iloc[0].tolist()
untereliste=pd.Series(untereliste).fillna("").tolist()
newlist=[]
for index in range(len(obereliste)):
    if str(obereliste[index]).startswith("Unnamed"):
        element=str(obereliste[index-1])+"_"+str(untereliste[index])
    else:
        element=str(obereliste[index])+"_"+str(untereliste[index])
    newlist.append(element)
data.columns=newlist
data
```

```python
gemeinde_data_2014_amtsang=data.dropna(subset=["Gemeindenummer_"])
gemeinde_data_2014_amtsang["AGS"]="120"+gemeinde_data_2014_amtsang["Gemeindenummer_"].str[:2]+gemeinde_data_2014_amtsang["Gemeindenummer_"].str[-3:]
gemeinde_data_2014_amtsang
```

```python
gemeinde_data_2014=pd.concat([gemeinde_data_2014_amtsfrei,gemeinde_data_2014_amtsang])
gemeinde_data_2014.to_csv("gemeindedaten_2014.csv",index=False)
```

```python
gemeinde_data_2019
```

```python
wahl2019_2014=pd.merge(gemeinde_data_2019,gemeinde_data_2014,on="AGS",suffixes=["_2019","_2014"],how="outer",indicator=True)
```

```python
#Hornow-Wadelsdorf fehlt da eingemeindet
```

```python
wahl2019_2014["diff_afd"]=pd.to_numeric(wahl2019_2014["AfD_%_2019"])-pd.to_numeric(wahl2019_2014["AfD_%_2014"])
wahl2019_2014.sort_values("diff_afd")
```

```python
wahl2019_2014.to_csv("election2014_2019.csv",index=False)
```

```python

kreis_data_2014_base=pd.read_csv("gemeindedaten_2014.csv",dtype={"AGS":str})
kreis_data_2014_base["AGS_K"]=kreis_data_2014_base["AGS"].str[:5]
kreis_data_2014_summed=kreis_data_2014_base.groupby("AGS_K").sum().reset_index()
kreis_data_2014_summed
```

```python
for i in kreis_data_2014_summed.columns.tolist():
    if(i.endswith("Anzahl")):
        partyname=(i[:-7])
        kreis_data_2014_summed[(partyname+"_%")]=kreis_data_2014_summed[(partyname+"_Anzahl")]/kreis_data_2014_summed["Wähler_Anzahl_Anzahl"]*100
```

```python
kreis_data_2014_summed
```

```python
data=pd.read_excel("../RawData/DL_BB_2_LT2019 (1).xlsx", sheet_name="Brandenburg_Landtagswahl_A_2")
data=data.rename(columns={"Unnamed: 3":"Name"})
obereliste=data.columns.tolist()
untereliste=data.iloc[0].tolist()
untereliste=pd.Series(untereliste).fillna("").tolist()
data[data.columns[2]].unique()
newlist=[]
for index in range(len(obereliste)):
    if str(obereliste[index]).startswith("Unnamed"):
        element=str(obereliste[index-1])+"_"+str(untereliste[index])
    else:
        element=str(obereliste[index])+"_"+str(untereliste[index])
    newlist.append(element)
data.columns=newlist
data=data.dropna(subset=[data.columns[2]])

landkreis_data_2019=data[(data[data.columns[2]].str.startswith("Landkreis"))]
landkreis_data_2019["AGS_K"]=landkreis_data_2019[gemeinde_data_2019.columns[1]].str[:5]
landkreis_data_2019
```

```python
landkreis2019_2014=pd.merge(landkreis_data_2019,kreis_data_2014_summed,on="AGS_K",suffixes=["_2019","_2014"],how="outer",indicator=True)
landkreis2019_2014
```

```python
landkreis2019_2014["diff_afd"]=pd.to_numeric(landkreis2019_2014["AfD_%_2019"])-pd.to_numeric(landkreis2019_2014["AfD_%_2014"])
landkreis2019_2014.sort_values("diff_afd")
```
