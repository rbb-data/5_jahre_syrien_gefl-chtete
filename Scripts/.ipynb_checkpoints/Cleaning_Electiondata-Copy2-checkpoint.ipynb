import pandas as pd
import altair as alt
import geopandas as gpd

```python
data=pd.read_csv("../CleanedData/auslaender_deutsche_gemeinde.csv")
data=data.rename(columns={"Unnamed: 0":"Stichtag","Unnamed: 1":"AGS_K","Unnamed: 2":"Name"})
data=data.dropna()
data
```

```python
auslaender_2016=data[data["date"]=="2018-12-31"]
auslaender_2016
```

```python
auslaender_2013=data[data["date"]=="2013-12-31"]
auslaender_2013
```

```python
diff_auslaender=pd.merge(auslaender_2013,auslaender_2016,on="AGS", suffixes=["_2013","_2016"])
diff_auslaender["diff_auslaender"]=diff_auslaender["Auslaender_Anteil_2016"]-diff_auslaender["Auslaender_Anteil_2013"]
diff_auslaender["diff_auslaender_rel"]=diff_auslaender["diff_auslaender"]/diff_auslaender["Auslaender_Anteil_2013"]

diff_auslaender.sort_values(by="diff_auslaender_rel")
```

```python
gemeinde2019_2014=pd.read_csv("election2014_2019.csv",dtype={"AGS":str})
gemeinde2019_2014
```

```python
afd_gemeinde=gemeinde2019_2014[["AGS","Name_","diff_afd","AfD_%_2019"]]
afd_gemeinde["AGS"]=afd_gemeinde["AGS"].astype(str)
afd_gemeinde
```

```python
correl=pd.merge(diff_auslaender,afd_gemeinde,left_on="AGS",right_on="AGS",how="inner")
correl.sort_values(by="diff_auslaender_rel")
```

```python
import scipy.stats as stats
import scipy
```

```python
charts=alt.Chart().mark_point().encode(
    x='diff_auslaender',
    y='diff_afd',
     tooltip="Name_",
)
bakedchart=charts + charts.transform_regression('diff_auslaender','diff_afd').mark_line()


result=(scipy.stats.linregress(correl["diff_auslaender"], correl["diff_afd"]))
print(result.pvalue)
print(result.rvalue**2)

combined = alt.layer(charts + charts.transform_regression('diff_auslaender','diff_afd').mark_line(), data=correl)
combined.display()
```

```python
reduced_correl=correl.sort_values(by="diff_auslaender_rel").reset_index(drop=True).iloc[20:-20].sort_values(by="diff_afd").reset_index(drop=True).iloc[20:-20]
```

```python
charts=alt.Chart().mark_point().encode(
    x='diff_auslaender_rel',
    y='diff_afd',
     tooltip="Name_",
)
bakedchart=charts + charts.transform_regression('diff_auslaender_rel','diff_afd').mark_line()


result=(scipy.stats.linregress(reduced_correl["diff_auslaender_rel"], reduced_correl["diff_afd"]))
print(result.pvalue)
print(result.rvalue**2)

combined = alt.layer(charts + charts.transform_regression('diff_auslaender_rel','diff_afd').mark_line(), data=reduced_correl)
combined.display()
```

```python
charts=alt.Chart().mark_point().encode(
    x='Auslaender_Anteil_2016',
    y='AfD_%_2019',
     tooltip="Name_",
)
bakedchart=charts + charts.transform_regression('Auslaender_Anteil_2016','AfD_%_2019').mark_line()


result=(scipy.stats.linregress(reduced_correl["Auslaender_Anteil_2016"], reduced_correl["AfD_%_2019"]))
print(result.pvalue)
print(result.rvalue**2)

combined = alt.layer(charts + charts.transform_regression('Auslaender_Anteil_2016','AfD_%_2019').mark_line(), data=reduced_correl)
combined.display()
```
