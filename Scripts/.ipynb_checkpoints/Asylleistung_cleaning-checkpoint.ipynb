```python
first_element=""
second_element=""
third_element=""
column_list=[]
entry=""
for i in range(len(data.iloc[1,:].tolist())):
    print(i)
    if(pd.isnull(data.iloc[1,:].tolist()[i])):
        pass
    else:
        first_element=data.iloc[1,:].tolist()[i]
    if(pd.isnull(data.iloc[2,:].tolist()[i])):
        pass
    else:
        second_element=data.iloc[2,:].tolist()[i]
    if(pd.isnull(data.iloc[3,:].tolist()[i])):
        pass
    else:
        third_element=data.iloc[3,:].tolist()[i]
    if (first_element+"_"+second_element+"_"+third_element==entry):
        entry=first_element+"_"+second_element+"_"+third_element+"_"+str(i)
    else:
        entry=first_element+"_"+second_element+"_"+third_element
    column_list.append(entry)
```

```python
datadropped=data.dropna(subset=[data.columns[0]])
indexlist=datadropped[datadropped[datadropped.columns[0]].str.startswith("31.12.")].index.tolist()
datadropped[datadropped[datadropped.columns[0]].str.startswith("31.12.")]
```

```python
indexlist
```

```python
firstdf=datadropped.loc[indexlist[0]:(indexlist[1]-1)]
firstdf.columns=column_list
firstdf["Datum"]=firstdf.iloc[0,0]
firstdf=firstdf[firstdf[firstdf.columns[0]].str.len()==5]
firstdf
```

```python
mergeddf=firstdf
for i in range(1,len(indexlist)-1):
    nextdf=datadropped.loc[indexlist[i]:(indexlist[i+1]-1)]
    nextdf.columns=column_list
    nextdf["Datum"]=nextdf.iloc[0,0]
    nextdf=nextdf[nextdf[nextdf.columns[0]].str.len()==5]
    mergeddf=pd.concat([mergeddf,nextdf])
mergeddf
```

```python
mergeddf["Anteil_ausländische_Sozialhilfeempfaenger"]=pd.to_numeric(mergeddf["Empfänger(innen) von Sozialhilfe_Ausländer(innen)_"],errors="coerce")/pd.to_numeric(mergeddf["Empfänger(innen) von Sozialhilfe__"],errors="coerce")
mergeddf
```

```python
mergeddf["Anteil_ausländische_Sozialhilfeempfaenger_außerhalb_Einrichtugen"]=pd.to_numeric(mergeddf["Empfänger(innen) von Sozialhilfe_ausländische Empfänger außerhalb von Einrichtungen_"],errors="coerce")/pd.to_numeric(mergeddf["Empfänger(innen) von Sozialhilfe_außerhalb von Einrichtungen_"],errors="coerce")
mergeddf
```
