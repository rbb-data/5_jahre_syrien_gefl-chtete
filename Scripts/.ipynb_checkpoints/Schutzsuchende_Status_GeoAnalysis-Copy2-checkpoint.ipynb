```python
data=pd.read_csv("../CleanedData/schutzsuchende_nations_kreis.csv",dtype={"AGS":str})
data
```

```python
since2016=data[data["Stichtag"]!="31.12.2013"]
since2016=since2016[since2016["Stichtag"]!="31.12.2014"]
since2016=since2016[since2016["Stichtag"]!="31.12.2015"]

since2016=since2016[since2016["AGS"].str[:2].isin(["11","12"])]
since2016_syrien_data=since2016[since2016["Herkunft"]=="Syrien"]
since2016_syrien_data[["Stichtag","Stadtname","Schutzsuchende_alle_Status"]]
```

```python
since2016_syrien_data_pivoted=since2016_syrien_data.pivot(index='Stadtname', columns='Stichtag', values='Schutzsuchende_alle_Status').reset_index().sort_values(by="31.12.2019",ascending=False)
since2016_syrien_data_pivoted=since2016_syrien_data_pivoted.iloc[:-1]
since2016_syrien_data_pivoted["Stadtname"]=since2016_syrien_data_pivoted["Stadtname"].str.replace(", kreisfreie Stadt","")
since2016_syrien_data_pivoted["Stadtname"]=since2016_syrien_data_pivoted["Stadtname"].str.replace(", Landkreis","")

since2016_syrien_data_pivoted.to_csv("../ChartData/since2016_syrien_data_pivoted.csv",index=False)
since2016_syrien_data_pivoted
```

```python
before2015=data[data["Stichtag"]=="31.12.2013"]
before2015=before2015[before2015["AGS"].str[:2].isin(["11","12"])]
before2015_syrien_data=before2015[before2015["Herkunft"]=="Syrien"]
before2015_syrien_data
```

```python
data=data[data["AGS"].str[:2].isin(["11","12"])]
syrien_data=data[data["Herkunft"]=="Syrien"]
syrien_data_max=syrien_data[syrien_data["Stichtag"].str[-2:].isin(["19"])].sort_values('Schutzsuchende_alle_Status', ascending=False).drop_duplicates(['AGS'])
syrien_data_max=syrien_data_max[["AGS","Stadtname","Schutzsuchende_alle_Status","Stichtag"]]
syrien_data_min=syrien_data[syrien_data["Stichtag"].str[-2:].isin(["16","15"])].sort_values('Schutzsuchende_alle_Status', ascending=False).drop_duplicates(['AGS'])
syrien_data_min
```

```python
syrien_data_min=syrien_data_min[["AGS","Stadtname","Schutzsuchende_alle_Status","Stichtag"]]
syrien_data_max_min=pd.merge(syrien_data_max,syrien_data_min,on=["AGS","Stadtname"],suffixes=["_max","_min"])
syrien_data_max_min["diff"]=syrien_data_max_min["Schutzsuchende_alle_Status_max"]-syrien_data_max_min["Schutzsuchende_alle_Status_min"]
zuzug=syrien_data_max_min[(syrien_data_max_min["Stichtag_max"].str[-2:])>(syrien_data_max_min["Stichtag_min"].str[-2:])]

syrien_data_max_min["diff"]=syrien_data_max_min[(syrien_data_max_min["Stichtag_max"].str[-2:])<(syrien_data_max_min["Stichtag_min"].str[-2:])]["diff"]*-1
wegzug=syrien_data_max_min.dropna()
wegzug
syrien_data_max_min=pd.concat([zuzug,wegzug])
syrien_data_max_min.sort_values("diff")
```

```python
syrien_data_max_min.to_csv("../ChartData/Syrerinnen_min_max.csv", index=False)
```

```python
#import matplotlib.pyplot as plt
country_shapes = kreise.merge(syrien_data_max_min, on='AGS')

country_shapes
color_map = plt.cm.get_cmap('Blues')

country_shapes=country_shapes.to_crs({'init': 'epsg:25832'})

mapp=country_shapes.plot("diff",cmap=color_map,edgecolor='black',scheme='fisherjenks',antialiased=False,legend=True,linewidth=0.5,figsize=(18, 16)).axis('off')
#mapp
```

```python
# Iterate through continents
filtered=data[(data["Herkunft"]==country)]
country_shapes = kreise.merge(filtered, on='AGS')
country_shapes=country_shapes.to_crs({'init': 'epsg:25832'})
jahre =country_shapes.groupby('Stichtag')

plt.figure()
```

```python
for i, (years, sub_df) in enumerate(jahre):
    new_df=(sub_df[["AGS",'geometry','Insgesamt']])
    ax = plt.subplot(3, 3, i + 1) # nrows, ncols, axes position
    # plot the continent on these axes
    new_df.plot("Insgesamt",ax=ax)
    # set the title
    ax.set_title(years)
    # set the aspect
    # adjustable datalim ensure that the plots have the same axes size
    ax.set_aspect('equal', adjustable='datalim')

plt.tight_layout()
plt.show()
```

```python
jahre
for i, (continent_name, continent_gdf) in enumerate(jahre):
    # create subplot axes in a 3x3 grid
    ax = plt.subplot(3, 3, i + 1) # nrows, ncols, axes position
    # plot the continent on these axes
    continent_gdf.plot("Insgesamt",ax=ax)
    # set the title
    ax.set_title(continent_name)
    # set the aspect
    # adjustable datalim ensure that the plots have the same axes size
    ax.set_aspect('equal', adjustable='datalim')

plt.tight_layout()
plt.show()
```
