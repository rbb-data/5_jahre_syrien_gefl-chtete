```python
import pandas as pd
import altair as alt
import geopandas as gpd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("../CleanedData/auslaender_status_aenderungen.csv",dtype={"AGS":str})
data.columns
```

```python
data["Insgesamt"]=data['Freizügigkeit nach EU-Recht_männlich',
       'Freizügigkeit nach EU-Recht_weiblich',
       'V.d.Erford.e.Aufenthaltstit.befreit, heimatl.Ausl._männlich',
       'V.d.Erford.e.Aufenthaltstit.befreit, heimatl.Ausl._weiblich',
       'Unbefristeter Aufenthaltstitel_männlich',
       'Unbefristeter Aufenthaltstitel_weiblich',
       'Befristeter Aufenthaltstitel_männlich',
       'Befristeter Aufenthaltstitel_weiblich','Antrag auf einen Aufenthaltstitel gestellt_männlich',
       'Antrag auf einen Aufenthaltstitel gestellt_weiblich',
       'Duldung_männlich', 'Duldung_weiblich',
       'Aufenthaltsgestattung_männlich', 'Aufenthaltsgestattung_weiblich',
       'Ohne Aufenthaltstitel, Duldung oder Gestattung_männlich',
       'Ohne Aufenthaltstitel, Duldung oder Gestattung_weiblich'].sum(axis=1)
data
```

```python
kreise=gpd.read_file("../../kreisebene/vg1000-ew_ebenen_3112/VG1000_KRS.shp")
kreise
```

```python
# Iterate through continents
filtered=data[(data["Herkunft"]==country)]
country_shapes = kreise.merge(filtered, on='AGS')
country_shapes=country_shapes.to_crs({'init': 'epsg:25832'})
jahre =country_shapes.groupby('Stichtag')

plt.figure()
```

```python
for i, (years, sub_df) in enumerate(jahre):
    new_df=(sub_df[["AGS",'geometry','Insgesamt']])
    ax = plt.subplot(3, 3, i + 1) # nrows, ncols, axes position
    # plot the continent on these axes
    new_df.plot("Insgesamt",ax=ax)
    # set the title
    ax.set_title(years)
    # set the aspect
    # adjustable datalim ensure that the plots have the same axes size
    ax.set_aspect('equal', adjustable='datalim')

plt.tight_layout()
plt.show()
```

```python
jahre
for i, (continent_name, continent_gdf) in enumerate(jahre):
    # create subplot axes in a 3x3 grid
    ax = plt.subplot(3, 3, i + 1) # nrows, ncols, axes position
    # plot the continent on these axes
    continent_gdf.plot("Insgesamt",ax=ax)
    # set the title
    ax.set_title(continent_name)
    # set the aspect
    # adjustable datalim ensure that the plots have the same axes size
    ax.set_aspect('equal', adjustable='datalim')

plt.tight_layout()
plt.show()
```
