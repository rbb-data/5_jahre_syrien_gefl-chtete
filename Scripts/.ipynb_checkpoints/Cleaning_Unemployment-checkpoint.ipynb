```python
import altair as alt
```

```python
import pandas as pd
import glob
```

```python
listoffiles=glob.glob("../RawData/arbeit_flucht/*")
listoffiles=sorted(listoffiles)
listoffiles.reverse()
listoffiles[0]
```

```python
filename=listoffiles[0]
data=pd.read_excel(listoffiles[0], sheet_name="4")
obereliste=data.iloc[7,:].fillna("Unnamed").tolist()
untereliste=data.iloc[9,:].fillna("").tolist()
newlist=[]
for index in range(len(obereliste)):
    if str(obereliste[index]).startswith("Unnamed"):
        element=str(obereliste[index-1])+"_"+str(untereliste[index])
    else:
        element=str(obereliste[index])+"_"+str(untereliste[index])
    element=element.replace("\n","")
    newlist.append(element)
data.columns=newlist
data["AGS"]=data[data.columns[0]].str[:5]
datawithoutna=data.dropna(subset=["AGS"])
kreise_data=datawithoutna[datawithoutna["AGS"].str.isdigit()]
kreise_data["date"]=filename[49:55]
base_data=kreise_data
base_data
```

```python
base_data.columns
```

```python
for filename in listoffiles[1:35]:
    print(filename)
    data=pd.read_excel(filename, sheet_name="4")
    data.columns=newlist
    data["AGS"]=data[data.columns[0]].str[:5]
    datawithoutna=data.dropna(subset=["AGS"])
    kreise_data=datawithoutna[datawithoutna["AGS"].str.isdigit()]
    kreise_data["date"]=filename[49:55]
    base_data=pd.concat([base_data,kreise_data])
```

```python
base_data
```

```python
len(base_data["AGS"].unique())
```

```python
base_data_no_dup=base_data.drop_duplicates()
base_data_no_dup["datum"]=pd.to_datetime(base_data_no_dup["date"],format='%Y%m')
base_data_no_dup
```

```python
filename=listoffiles[35]
print(filename)

data=pd.read_excel(filename, sheet_name="4")
data=data.dropna(axis=1,how="all")
obereliste=data.iloc[7,:].fillna("Unnamed").tolist()
untereliste=data.iloc[9,:].fillna("").tolist()
newlist=[]
for index in range(len(obereliste)):
    if str(obereliste[index]).startswith("Unnamed"):
        element=str(obereliste[index-1])+"_"+str(untereliste[index])
    else:
        element=str(obereliste[index])+"_"+str(untereliste[index])
    element=element.replace("\n","")
    newlist.append(element)
data.columns=newlist
data["AGS"]=data[data.columns[0]].str[:5]
datawithoutna=data.dropna(subset=["AGS"])
kreise_data=datawithoutna[datawithoutna["AGS"].str.isdigit()]
kreise_data["date"]=filename[49:55]
base_data_extra=kreise_data
base_data_extra["datum"]=pd.to_datetime(base_data_extra["date"],format='%Y%m')

base_data_extra.columns=base_data_no_dup.columns
base_data_extra
```

```python
until2017=pd.concat([base_data_no_dup,base_data_extra])
```

```python
filename=listoffiles[36]
print(filename)

data=pd.read_excel(filename, sheet_name="3")
data=data.dropna(axis=1,how="all")
obereliste=data.iloc[7,:].fillna("Unnamed").tolist()
untereliste=data.iloc[9,:].fillna("").tolist()
newlist=[]
for index in range(len(obereliste)):
    if str(obereliste[index]).startswith("Unnamed"):
        element=str(obereliste[index-1])+"_"+str(untereliste[index])
    else:
        element=str(obereliste[index])+"_"+str(untereliste[index])
    element=element.replace("\n","")
    newlist.append(element)
data.columns=newlist
data["AGS"]=data[data.columns[0]].str[:5]
datawithoutna=data.dropna(subset=["AGS"])
kreise_data=datawithoutna[datawithoutna["AGS"].str.isdigit()]
kreise_data["date"]=filename[49:55]
base_data=kreise_data
base_data
```

```python
for filename in listoffiles[37:]:
    print(filename)
    data=pd.read_excel(filename, sheet_name="3")
    print(data)
    #data=data.dropna(axis=1,how="all")
    data.columns=newlist

    data["AGS"]=data[data.columns[0]].str[:5]
    datawithoutna=data.dropna(subset=["AGS"])
    kreise_data=datawithoutna[datawithoutna["AGS"].str.isdigit()]
    kreise_data["date"]=filename[49:55]
    print(kreise_data)
    base_data=pd.concat([base_data,kreise_data])
```

```python
base_data["datum"]=pd.to_datetime(base_data["date"],format='%Y%m')

base_data.columns=until2017.columns

allyears=pd.concat([until2017,base_data])
```

```python
i="11000"
lk_data=allyears[allyears["AGS"]==i]
lk_data=lk_data.iloc[1:]
lk_data.sort_values("datum")
```

```python
lk_data[lk_data.columns[3]]=pd.to_numeric(lk_data[lk_data.columns[3]])
```

```python
alt.Chart(lk_data).mark_line().encode(
    x='datum',
    y="ratio"
)
```

```python
allyears["ratio"]=pd.to_numeric(allyears[allyears.columns[2]],errors="coerce")/pd.to_numeric(allyears[allyears.columns[1]],errors="coerce")*100
allyears
```
