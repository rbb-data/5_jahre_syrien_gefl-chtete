```python
import pandas as pd
```

```python
data=pd.read_csv("../CleanedData/auslaender_aufenthaltstitel_kreis.csv",dtype={"AGS":str})
data.columns
```

```python
newdf=pd.DataFrame()
for i in data["Herkunft"].unique().tolist():
    print(i)
    for j in data["AGS"].unique().tolist():
        onecountry=data[data["Herkunft"]==i].sort_values(by=["AGS","Stichtag"])
        changes=onecountry[onecountry["AGS"]==j][onecountry.columns[4:]].fillna(0)-onecountry[onecountry["AGS"]==j][onecountry.columns[4:]].shift(1).fillna(0)
        frontdf=onecountry[onecountry["AGS"]==j][onecountry.columns[:4]]
        addingdf=pd.concat([frontdf,changes],axis=1)
        newdf=pd.concat([newdf,addingdf],axis=0)
```

```python
newdf
```

```python
newdf.to_csv("status_aenderungen.csv",index=False)
```

```python
newdf["Summe_alle_Aufenthaltstitel"]=newdf[4:].sum(axis=1)
```
